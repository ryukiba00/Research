## 0.關於TFF

類似TF框架，TFF實際運行的代碼並不是Python，而是通過Python代碼來編寫運算邏輯，實際上是編譯成另一種語言去執行。Why？

因為許多設備（例如手機、傳感器）是很難有Python的運行環境的，更不可能去安裝幾百Mb的TensorFlow框架，那麼在這些設備上執行Python代碼的難度是非常大的。

所以，學習TFF框架時，應該把它當成一門新的語言進行學習：這裡有新的數據類型、新的變量聲明方式、新的函數定義方式、新的執行方法等。

兩個組成元件

1.   Federated Learning API: 供了一組高階接口，使開發者能夠將包含的聯合訓練和評估實現應用於現有的TensorFlow 模型。
2.   Federated Core API: 一組較低階接口，通過在強類型函數式編程環境中結合使用TensorFlow與分佈式通信運算符，簡潔地表達新的聯合算法。
